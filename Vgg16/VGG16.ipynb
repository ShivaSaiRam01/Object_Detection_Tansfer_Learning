```python
import tensorflow as tf

# load the model
model = tf.keras.applications.VGG16()

layer_names = [layer.name for layer in model.layers]

print(layer_names)
print(len(layer_names))
```

```python
num_params = sum([tf.keras.backend.count_params(w) for w in model.trainable_weights])

print('Number of parameters',num_params)
```

```python
num_classes = model.output_shape[-1]

print('Number of classes : ',num_classes)
```

```python
class_names_path = tf.keras.utils.get_file('ImageNetLabels.txt','https://storage.googleapis.com/download.tensorflow.org/data/ImageNetLabels.txt')
with open (class_names_path, 'r') as f:
  class_names = f.read().splitlines()
print('class names :',class_names)
```

```python
import tensorflow as tf
from tensorflow.keras.utils import plot_model

plot_model(model,to_file = 'model.png',show_shapes=True,show_layer_names=True)
```

```python
from keras .preprocessing.image import ImageDataGenerator
from keras.preprocessing import image

import numpy as np
img_path = '/content/tops.jpg'

img = image.load_img(img_path ,target_size = (224,224))
img
```

```python
img_array = image.img_to_array(img)
img_array.shape
```

```python
img_array_expanded_dims = np.expand_dims(img_array, axis=0)
img_array_expanded_dims.shape
```

```python
predictions = model.predict(img_array_expanded_dims)
predictions.shape
```

```python
from keras.applications import imagenet_utils
results = imagenet_utils.decode_predictions(predictions)
results
```
